#### Kullanıcı Tanımlı Fonksiyonlar
* Kullanıcı tarafından tanımlanır.
* Kullanıcı tanımlı bir işleve herhangi bir ad verebilirsiniz, ancak Python anahtar sözcüklerini işlev adı olarak kullanamazsınız. 

* Bir dairenin çevresini ve alanını bulmak için bir dizi işlem yapalım.

```python
pi_sayisi = 3.14
yaricap = 6
# alan
daire_alan = pi_sayisi*(yaricap**2) # yaricap**2 ifadesi yaricapi karesidir 
daire_alan
```

```python
# çevre
cember_cevresi = 2*pi_sayisi*yaricap
cember_cevresi
```

* Dairenin alanını hesaplamak için fonksiyon tanımlayalım.

```python
# def = definition
# daireAlanHesapla = fonksiyon ismi
# r = girdi parametresi
def daireAlanHesapla(r):
    """
        Dairenin alanını hesaplar
        Girdi = Yarı Çap
        Çıktı = Daire Alanı
    """
    pi = 3.14
    daire_alanı = pi*(r**2)
    return daire_alanı

Alan = daireAlanHesapla(5)
print(Alan)
Alan = daireAlanHesapla(5.5)
print(Alan)
Alan = daireAlanHesapla(5.7)
print(Alan)
Alan = daireAlanHesapla(5.8)
print(Alan)
Alan = daireAlanHesapla(15)
print(Alan)
```

* Çemberin çevresini hesaplamak için fonksiyon tanımlayalım.

```python
def cemberCevreHesapla(r):
    """
        Çemberin cevresini hesaplar
        Girdi = Yarı Çap
        Çıktı = Çemberin Çevresi
    """
    pi = 3.14
    cember_cevre = 2*pi*r
    return cember_cevre
cemberCevreHesapla(4)
```

#### Lambda Fonksiyonu
* İleri seviye bir fonksiyon çeşididir.
* Lambda işlevi, küçük anonim bir işlevdir.
* Bir lambda işlevi herhangi bir sayıda argüman alabilir, ancak yalnızca bir ifadeye sahip olabilir.

```python
# çarpma işlemi yapan normal fonksiyon
def carpma(x,y,z,t,w):
    return (x*y*z)-(t*w)

carpma(2,3,4,5,6)
```

```python
# aynı işlemi lambda fonksiyon ile yapmak
fonksiyon_lambda = lambda x, y, z, w, t : (x*y*z)-(t*w)
fonksiyon_lambda(2,3,4,5,6)
```

* Lambda'nın gücü, onları başka bir fonksiyon içinde anonim bir fonksiyon olarak kullandığınızda daha iyi gösterilir.

```python
def fonksiyonLambda(n):
  return lambda a : a * n

ana_fonksiyon = fonksiyonLambda(5)
ana_fonksiyon(7)
```

### Döngüler: For
* Bir dizi üzerinde yineleme yapmak için for döngüsü kullanılır. Bu dizi bir liste, bir sözlük, veya bir metindir.
* For döngüsü ile bir liste, tuple, set vb. İçindeki her öğe için bir kez olmak üzere bir dizi deyimi çalıştırabiliriz.

```python
# range 2'den 13'e kadar değerleri i parametresine her iterasyonda yazar
for i in range(2,13): # (i = 2, i < 13, i++)
    print(i)
    
print(10*i)
```

```python
for i in "ankara.!012 konya":
    print(i)
print(i + "Deneme")
```

```python
metin = "ankara konya antalya"
metin_ayrık = metin.split() # boşluğa göre ayırır
metin_ayrık
```

```python
for i in metin_ayrık: 
    print(i)
```

```python
liste = [1,4,5,6,8,3,3,4,65,6]
# for döngüsü ile toplama
count = 0
for c in liste:
    count = count + c

print(count)
```

### Döngüler: While
* While döngüsü ile, bir koşul doğru olduğu sürece bir dizi ifadeyi çalıştırabiliriz.


```python
i = 5
while(i <= 8):
    print(i)
    i = i + 1
```

```python
# while ile toplama
liste = [1,4,5,6,8,3,3,4,67]

sinir = len(liste)   
indis = 0
toplam = 0
while(indis < sinir):
    toplam = toplam + liste[indis]
    indis = indis + 1 
toplam
```

### Liste Kavrama
* Liste kavrama veri analizinde veriyi işlemek için kullanılan etkili bir yöntemdir.

```python
# veriyi oluştur
import pandas as pd
sozluk = {"isim": ["ali","veli","kenan","murat","ayse","hilal"],
              "yas" : [15,16,17,33,45,66],
              "sehir": ["İzmir","Ankara","Konya","Ankara","Ankara","Antalya"]} 
veri = pd.DataFrame(sozluk)
veri
```

```python
# ortalama yasi bulalım
ortalama_yas = veri.yas.mean()
# np.mean(veri.yas) yöntemini de kullanabilirdik
ortalama_yas
```

* Ortalama yaşı bulduktan sonra verimizi büyük ve küçük olmak üzere ikiye ayıralım. Ortalama yaştan büyük yaşa sahip olan örneklere büyük, diğer örneklere ise küçük etiketini koyalım.

```python
veri["YAS_GRUBU"] = ["kucuk" if ortalama_yas > i else "buyuk" for i in veri.yas]
veri
```

* Başka bir örnek ise sütunların hepsini küçük harf yapmak olabilir.
* Böylece YAS_GRUBU küçük harf olacaktır.

```python
veri.columns = [ i.lower() for i in veri.columns] 
veri
```

* Liste kavramı yöntemi yerine daha önceden öğrendiğimiz lambda fonksiyonunu kullanabiliriz.

```python
veri = veri.assign(yas_5_yil_sonra = lambda x: (x['yas'] + 5))
veri
```

* apply metodunu liste kavrama ve lambda fonksiyonu yerine kullanabiliriz.

```python
def yas10YilSonra(age):
    return age + 10
```

```python
veri["yas_10_yil_sonra"] = veri.yas.apply(yas10YilSonra)
veri
```

### Histogram Grafiği
* Histogram, gruplandırılmış bir veri dağılımının sütun grafiğiyle gösterimidir. Diğer bir ifadeyle, tekrarlı sayılardan oluşan verilerin, uygulanan işlemlerden sonra önce tabloya, tablodan yararlanarak grafiğe aktarılması, yani veri gruplarının grafiğinin dikdörtgen sütunlar halinde gösterilmesidir. 

```python
# Kütüphaneleri içe aktar ve veriyi yükle
import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("iris.csv")
veri.head()
```

```python
# şimdi sepal length sütununu id'ye göre çizdirelim
plt.figure() # figür oluştur - dış çerçeve
plt.hist(veri["SepalLengthCm"], color = "green", alpha = 0.5, bins = 20, label = "Alt Yaprak Uzunluk") # alpha = saydamlık, bins = bar sayısı
plt.hist(veri["PetalLengthCm"], color = "orange", alpha = 0.5, bins = 20, label = "Üst Yaprak Uzunluk") # alpha = saydamlık, bins = bar sayısı
plt.title('Alt Yaprak Uzunluğu ve Üst Yaprak Uzunluğu - Histogram')
plt.xlabel("Cm")
plt.ylabel("Frekans")
plt.legend() # label'ların görünebilmesi için
plt.show()
```

### Çubuk Grafiği

### Alt Plan Grafiği
* Alt plan grafiklerinde birden fazla grafik tek bir figür/çerçeve üzerinde bulunur.

```python
# Kütüphaneleri içe aktar ve veriyi yükle
import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("iris.csv")
veri.head()
```

```python
fig, ax = plt.subplots(2, 1) # figür üzerine 2*1 lik bir tablo oluştur. bu tablo üzerine ax (eksen) ekl
fig.suptitle("Alt Yaprak Uzunluk ve Üst Yaprak Uzunluk", fontsize=14)
ax[0].plot(veri.Id,veri.SepalLengthCm,color="red",label= "SepalLengthCm")
ax[0].set_title('Alt Yaprak Uzunluk') # başlık
ax[0].set_xlabel("id") # x ekseni label
ax[0].set_ylabel("Cm") # y ekseni label

ax[1].plot(veri.Id,veri.PetalLengthCm,color="green",label= "PetalLengthCm")
ax[1].set_title('Üst Yaprak Uzunluk')
ax[1].set_xlabel("id")
ax[1].set_ylabel("Cm")

fig.tight_layout() # figür arası yerleşimi ayarla
fig.subplots_adjust(top=0.85)
plt.show()
```
