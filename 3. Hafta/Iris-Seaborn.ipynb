```python
# veri setini y√ºkleyelim
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

veri = pd.read_csv("iris.csv")
veri.head()
```

```python
# buraya kod yazacaƒüƒ±z
sns.countplot(x="Species", data=veri)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.displot(veri, x="SepalLengthCm", hue="Species")
plt.ylabel("Frekans")
plt.title("T√ºre G√∂re Sepal Length Histogramƒ±")
plt.show()
```

```python
sns.displot(veri, x="SepalLengthCm", col="Species", multiple="dodge")
plt.show()
```

```python
sns.displot(veri, x="SepalLengthCm", y="SepalWidthCm", kind="kde")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Width Histogram")
plt.show()
```

```python
sns.displot(veri, x="SepalLengthCm", y="SepalWidthCm", hue="Species", kind="kde")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Width Histogram")
plt.show()
```

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=veri)
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=veri)
plt.show()
```

```python
sns.boxplot(x="Species", y="SepalWidthCm", data=veri)
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalWidthCm", data=veri)
plt.show()
```

```python
sns.heatmap(veri.corr(), annot=True, linewidths=.5, fmt= '.1f')
plt.show()
```

```python
sns.violinplot(x="Species", y="SepalLengthCm", data=veri)
plt.show()
```

```python
# buraya kod yazacaƒüƒ±z
sns.jointplot(
    data=veri,
    x="SepalLengthCm", y="SepalWidthCm", hue="Species",
    kind="kde")
plt.show()
```

```python
# buraya kod yazacaƒüƒ±z
g = sns.JointGrid(data=veri, x="SepalLengthCm", y="SepalWidthCm")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
# buraya kod yazacaƒüƒ±z
sns.pairplot(veri)
plt.show()
```

```python
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
