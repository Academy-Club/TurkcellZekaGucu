```python
import pandas as pd 
  
data = pd.read_csv("nba.csv") 

data.head()
```

```python
data = data.fillna(method ='pad')

data.head()
```

```python
conversions = [30.48, 2.54]
data['newHeight'] = data['Height'].str.split('-').apply(pd.Series).astype(int).dot(conversions)
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

ax = data.plot.scatter(x='newHeight',y='Weight')
```

```python
ax = data.plot.scatter(x='newHeight',y='Salary')
```

```python
ax = data.plot.scatter(x='Age',y='Salary')
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x='newHeight',y='Weight', hue='Position', data=data)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs. Weight (Colored w.r.t Positions)')
plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x='Age',y='Salary', hue='Position', data=data)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary vs. Age (Colored w.r.t Positions)')
plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x='Age',y='Salary', hue='Team', data=data)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs. Weight (Colored w.r.t Teams)')
plt.show()
```

```python
sns.displot(data, x="newHeight", hue="Position")
plt.ylabel("Frequency")
plt.title("Height Histogram (categorized to Position)")
plt.show()
```

```python
sns.displot(data, x="newHeight", col="Position", multiple="dodge")
plt.show()
```

```python
sns.displot(data, x="Salary", hue="Position")
plt.ylabel("Frequency")
plt.title("Salary Histogram (categorized to Position)")
plt.show()
```

```python
sns.displot(data, x="Salary", col="Position", multiple="dodge")
plt.show()
```

```python
sns.barplot(x="Position", y="Salary", data=data)
plt.title("Salary based on Position")
plt.show()
```

```python
sns.barplot(x="Position", y="newHeight", data=data)
plt.title("Height based on Position")
plt.show()
```

```python
sns.barplot(x="Team", y="Salary", data=data)
plt.xticks(rotation = 90)
plt.title("Salary based on Team")
plt.show()
```

```python
sns.boxplot(x="Position", y="newHeight", data=data)
plt.show()
```

```python
sns.boxplot(x="Position", y="Salary", data=data)
plt.show()
```

```python
sns.boxplot(x="Team", y="Salary", data=data)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f')
plt.show()
```
